## Choropleth map

```python
working_fields = paygap.columns[2:]  # Assuming columns start from index 2

# Create a new column in the DataFrame to store the maximum value across working fields
paygap['Max_Value'] = paygap[working_fields].max(axis=1)
paygap['Max_Variable'] = paygap[working_fields].idxmax(axis=1) # Get the column name with the maximum value

# Group by country and year, then calculate the maximum value and its corresponding variable for each group
max_values = paygap.groupby(['Country', 'Year'])[['Max_Value', 'Max_Variable']].max().reset_index()

# Create an interactive map with a distinct color scale and hover template
fig = px.choropleth(max_values,
                    locations="Country",
                    locationmode='country names',
                    color="Max_Variable",
                    hover_data=["Max_Variable", "Max_Value"],
                    animation_frame="Year",
                    color_continuous_scale=px.colors.qualitative.Plotly
                    )

# Set a larger projection type (e.g., "natural earth")
fig.update_geos(projection_type="natural earth")

fig.show()
```

# Countries situation

```python
years = paygap['Year']
variables = paygap.columns[2:]  # Assuming columns start from index 2

app1 = dash.Dash(__name__)

app1.layout = html.Div([
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in paygap['Country'].unique()],
        value=paygap['Country'].unique()[0]
    ),
    dcc.Graph(id='grouped-bar-chart'),
    dcc.Slider(
        id='year-slider',
        min=min(years),
        max=max(years),
        value=min(years),
        marks={str(year): str(year) for year in years},
        step=None
    )
])

@app1.callback(
    Output('grouped-bar-chart', 'figure'),
    Input('country-dropdown', 'value'),
    Input('year-slider', 'value')
)
def update_grouped_bar_chart(selected_country, selected_year):
    filtered_data = paygap[(paygap['Country'] == selected_country) & (paygap['Year'] == selected_year)]
    trace_list = []
    for variable in variables:
        trace = go.Bar(x=[variable], y=[filtered_data[variable].values[0]], name=variable)
        trace_list.append(trace)

    return {
        'data': trace_list,
        'layout': go.Layout(title=f'Variable Values for {selected_country} in {selected_year}',
                            xaxis={'title': 'Variable'},
                            yaxis={'title': 'Value'},
                            barmode='group')
    }

if __name__ == '__main__':
    app1.run_server(debug=True, port=8050)
```

# Variables situation

## Variables dominance

```python
app2 = dash.Dash(__name__)

app2.layout = html.Div([
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in paygap['Year'].unique()],
        value=paygap['Year'].min()
    ),
    dcc.Dropdown(
        id='variable-dropdown',
        options=[{'label': variable, 'value': variable} for variable in paygap.columns[2:]],
        value=paygap.columns[2]
    ),
    dcc.Graph(
    id='interactive-map',
    style={'width': '100%', 'height': '600px'})
])

@app2.callback(
    Output('interactive-map', 'figure'),
    Input('year-dropdown', 'value'),
    Input('variable-dropdown', 'value')
)
def update_interactive_map(selected_year, selected_variable):
    filtered_data = paygap[paygap['Year'] == selected_year]

    fig = px.scatter_geo(
        filtered_data,
        locations='Country',
        locationmode='country names',
        hover_name='Country',
        size=filtered_data[selected_variable],
        color=filtered_data[selected_variable],
        color_continuous_scale='Viridis',
        title=f'Interactive Map: {selected_variable} in {selected_year}'
    )

    fig.update_geos(
        projection_type="natural earth",
        showcoastlines=True, coastlinecolor="Black",
        showland=True, landcolor="white",
        showcountries=True, countrycolor="Black", countrywidth=0.5
    )

    return fig

if __name__ == '__main__':
    app2.run_server(debug=True, port=8051)
```

## Line plot: variables trends

```python
app3 = dash.Dash(__name__)
app3.layout = html.Div([
    dcc.Dropdown(
        id='variable-dropdown',
        options=[{'label': column, 'value': column} for column in paygap.columns[2:]],
        value=paygap.columns[2]
    ),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in paygap['Country'].unique()],
        value=paygap['Country'].unique()[0]
    ),
    dcc.Graph(id='line-plot')
])

# Define callback to update the line plot based on dropdown values
@app3.callback(
    Output('line-plot', 'figure'),
    Input('variable-dropdown', 'value'),
    Input('country-dropdown', 'value')
)
def update_line_plot(selected_variable, selected_country):
    filtered_data = paygap[paygap['Country'] == selected_country]

    fig = px.line(filtered_data, x='Year', y=selected_variable,
                  labels={'Your_Variable_Column': 'Variable Value'},
                  markers=True)

    fig.update_layout(title=f'Change Over Time for {selected_country}',
                      xaxis_title='Year', yaxis_title='Variable Value')

    return fig

if __name__ == '__main__':
    app3.run_server(debug=True, port= 8052)
```

## Correlation between variables

```python
app4= dash.Dash(__name__)
app4.layout = html.Div([
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in paygap['Year'].unique()],
        value=paygap['Year'].min()
    ),
    dcc.Dropdown(
        id='x-variable-dropdown',
        options=[{'label': variable, 'value': variable} for variable in paygap.columns[2:]],
        value=paygap.columns[2]
    ),
    dcc.Dropdown(
        id='y-variable-dropdown',
        options=[{'label': variable, 'value': variable} for variable in paygap.columns[2:]],
        value=paygap.columns[3]
    ),
    dcc.Graph(id='interactive-3d-scatter')
])

@app4.callback(
    Output('interactive-3d-scatter', 'figure'),
    Input('year-dropdown', 'value'),
    Input('x-variable-dropdown', 'value'),
    Input('y-variable-dropdown', 'value')
)
def update_3d_scatter(selected_year, selected_x_variable, selected_y_variable):
    filtered_data = paygap[paygap['Year'] == selected_year]

    fig = px.scatter_3d(
        filtered_data,
        x=selected_x_variable,
        y=selected_y_variable,
        z='Country',
        color=selected_x_variable,
        size=selected_y_variable,
        text='Country',
        title=f'Interactive 3D Scatter Plot: {selected_x_variable} vs {selected_y_variable} in {selected_year}'
    )

    return fig

if __name__ == '__main__':
    app4.run_server(debug=True, port=8053)
```

```python
app5 = dash.Dash(__name__)

app5.layout = html.Div([
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in paygap['Year'].unique()],
        value=paygap['Year'].min()
    ),
    dcc.Dropdown(
        id='x-variable-dropdown',
        options=[{'label': variable, 'value': variable} for variable in paygap.columns[2:]],
        value=paygap.columns[2]
    ),
    dcc.Dropdown(
        id='y-variable-dropdown',
        options=[{'label': variable, 'value': variable} for variable in paygap.columns[2:]],
        value=paygap.columns[3]
    ),
    dcc.Graph(id='interactive-3d-scatter')
])

@app4.callback(
    Output('interactive-3d-scatter', 'figure'),
    Input('year-dropdown', 'value'),
    Input('x-variable-dropdown', 'value'),
    Input('y-variable-dropdown', 'value')
)
def update_3d_scatter(selected_year, selected_x_variable, selected_y_variable):
    filtered_data = paygap[paygap['Year'] == selected_year]

    fig = px.scatter_3d(
        filtered_data,
        x=selected_x_variable,
        y=selected_y_variable,
        z='Country',
        color=selected_x_variable,
        size=selected_y_variable,
        text='Country',
        title=f'Interactive 3D Scatter Plot: {selected_x_variable} vs {selected_y_variable} in {selected_year}'
    )
    return fig

if __name__ == '__main__':
    app5.run_server(debug=True, port=8054)
```

```python
app6 = dash.Dash(__name__)

# Data Preprocessing
data = paygap.drop(['Country', 'Year'], axis=1)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Create cluster labels using K-Means
num_clusters = 6
kmeans = KMeans(n_clusters=num_clusters)
clusters = kmeans.fit_predict(scaled_data)

# Add Cluster Labels to DataFrame
paygap['Cluster'] = clusters

# Create an interactive map using Plotly Express
def create_cluster_map(selected_year):
    filtered_data = paygap[paygap['Year'] == selected_year]
    fig = px.choropleth(
        filtered_data,
        locations='Country',
        locationmode='country names',
        color='Cluster',
        color_continuous_scale='rainbow',
        hover_name='Country',
        title=f'Cluster Map for Year {selected_year}',
    )

    tickvals = list(range(num_clusters))
    ticktext = [str(cluster) for cluster in tickvals]
    fig.update_layout(coloraxis_colorbar=dict(tickvals=tickvals, ticktext=ticktext))

    return fig

app6.layout = html.Div([
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': str(year), 'value': year} for year in paygap['Year'].unique()],
        value=paygap['Year'].min()
    ),
    dcc.Graph(id='cluster-map')
])

@app6.callback(
    Output('cluster-map', 'figure'),
    Input('year-dropdown', 'value')
)
def update_cluster_map(selected_year):
    return create_cluster_map(selected_year)

if __name__ == '__main__':
    app6.run_server(debug=True, port=8055)
```
